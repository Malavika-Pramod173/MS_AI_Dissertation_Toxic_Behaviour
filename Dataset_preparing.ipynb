```python
import numpy as np

def build_jigsaw_toxic(df):
    df = df.copy()

    toxicity_cols = ["toxic", "severe_toxic", "obscene", "threat", "insult", "identity_hate"]

    # Coarse label: toxic if any toxicity flag is 1
    df["coarse_label"] = np.where(df[toxicity_cols].sum(axis=1) > 0, "toxic", "non_toxic")

    # Store original human labels as a dict per row (subset)
    df["human_label_original"] = df[toxicity_cols].to_dict(orient="records")

    # Standardised columns
    df["text"] = df["comment_text"]
    df["dataset_source"] = "jigsaw_toxic"

    return df[["text", "human_label_original", "coarse_label", "dataset_source"]]


def build_jigsaw_bias(df):
    df = df.copy()

    # Coarse label based on 'target' (overall toxicity score)
    # You can adjust the threshold later if needed
    df["coarse_label"] = np.where(df["target"] >= 0.5, "toxic", "non_toxic")

    # Select a subset of useful label columns to keep
    label_cols = [
        "target",
        "toxicity",
        "severe_toxicity",
        "obscene",
        "identity_attack",
        "insult",
        "threat",
        "sexual_explicit",
    ]
    # Some columns may be missing depending on version, so intersect with existing
    label_cols = [c for c in label_cols if c in df.columns]

    df["human_label_original"] = df[label_cols].to_dict(orient="records")

    df["text"] = df["comment_text"]
    df["dataset_source"] = "jigsaw_bias"

    return df[["text", "human_label_original", "coarse_label", "dataset_source"]]


def build_cyberbullying(df):
    df = df.copy()

    # Coarse label: toxic if any cyberbullying type other than "not_cyberbullying"
    df["coarse_label"] = np.where(
        df["cyberbullying_type"].str.lower() == "not_cyberbullying",
        "non_toxic",
        "toxic",
    )

    df["human_label_original"] = df["cyberbullying_type"]
    df["text"] = df["tweet_text"]
    df["dataset_source"] = "cyberbullying_kaggle"

    return df[["text", "human_label_original", "coarse_label", "dataset_source"]]


# Build unified views
df_jigsaw_unified = build_jigsaw_toxic(df_jigsaw)
df_bias_unified = build_jigsaw_bias(df_bias)
df_cyb_unified = build_cyberbullying(df_cyb_kaggle)

# Combine into one big frame
df_all = pd.concat(
    [df_jigsaw_unified, df_bias_unified, df_cyb_unified],
    ignore_index=True
)

print("Unified Jigsaw Toxic:", df_jigsaw_unified.shape)
print("Unified Jigsaw Bias:", df_bias_unified.shape)
print("Unified Cyberbullying:", df_cyb_unified.shape)
print("Combined unified dataset:", df_all.shape)
```

```python
import numpy as np

# Define the 10 final thesis categories
TOXIC_CATEGORIES = [
    "sexual_harassment",
    "cyberbullying",
    "hate_speech",
    "abusive_violence",
    "other_toxic"
]

NONTOXIC_CATEGORIES = [
    "non_sexual",
    "non_cyberbullying",
    "motivational_non_hate",
    "non_abusive_violence",
    "generic_neutral"
]


def map_jigsaw_toxic_labels(row):
    labels = row["human_label_original"]

    # hate_speech
    if labels.get("identity_hate", 0) == 1:
        return "hate_speech"

    # abusive_violence
    if labels.get("threat", 0) == 1 or labels.get("insult", 0) == 1:
        return "abusive_violence"

    # sexual_harassment (weak signal, handled better by LLM if needed)
    if labels.get("obscene", 0) == 1 and labels.get("severe_toxic", 0) == 1:
        return "sexual_harassment"

    # other_toxic
    if row["coarse_label"] == "toxic":
        return "other_toxic"

    # generic_non-toxic
    return "generic_neutral"



def map_jigsaw_bias_labels(row):
    labels = row["human_label_original"]

    # sexual_harassment
    if labels.get("sexual_explicit", 0) >= 0.5:
        return "sexual_harassment"

    # hate_speech
    if labels.get("identity_attack", 0) >= 0.5:
        return "hate_speech"

    # abusive_violence
    if labels.get("threat", 0) >= 0.5 or labels.get("insult", 0) >= 0.5:
        return "abusive_violence"

    # other_toxic
    if row["coarse_label"] == "toxic":
        return "other_toxic"

    # general categories will need LLM refinement
    return None



def map_cyberbullying_labels(row):
    label = row["human_label_original"]

    if label != "not_cyberbullying":
        return "cyberbullying"
    else:
        return "non_cyberbullying"



def assign_initial_category(row):
    source = row["dataset_source"]

    if source == "jigsaw_toxic":
        return map_jigsaw_toxic_labels(row)

    if source == "jigsaw_bias":
        return map_jigsaw_bias_labels(row)

    if source == "cyberbullying_kaggle":
        return map_cyberbullying_labels(row)

    return None  # fallback



# Apply category mapping
df_all["initial_category"] = df_all.apply(assign_initial_category, axis=1)

# Mark entries that require LLM refinement
df_all["needs_llm"] = df_all["initial_category"].isna()

print("Category mapping complete.")
print(df_all["initial_category"].value_counts(dropna=False))
print("\nCounts of entries needing LLM:")
print(df_all["needs_llm"].value_counts())
```

```python
TARGET_PER_CLASS = 1000

HUMAN_ONLY_CATEGORIES = [
    "cyberbullying",
    "abusive_violence",
    "other_toxic",
    "hate_speech",
    "sexual_harassment",
    "generic_neutral",
    "non_cyberbullying",
]

samples = []

for cat in HUMAN_ONLY_CATEGORIES:
    df_cat = df_all[df_all["initial_category"] == cat]

    available = len(df_cat)
    if available < TARGET_PER_CLASS:
        raise ValueError(f"Not enough samples for category '{cat}': have {available}, need {TARGET_PER_CLASS}")

    df_sampled = df_cat.sample(n=TARGET_PER_CLASS, random_state=42)
    df_sampled = df_sampled.copy()
    df_sampled["final_category"] = cat
    df_sampled["label_source"] = "human_direct"

    samples.append(df_sampled)

df_human_final = pd.concat(samples, ignore_index=True)

print("Human-only final dataset shape:", df_human_final.shape)
print("Counts per final_category in human-only dataset:")
print(df_human_final["final_category"].value_counts())
print("\nCoarse label distribution in human-only dataset:")
print(df_human_final["coarse_label"].value_counts())
print("\nSources in human-only dataset:")
print(df_human_final["dataset_source"].value_counts())
```

```python
import os
import json
from groq import Groq
import getpass

# Use environment variable for API key security
api_key = os.getenv("GROQ_API_KEY", "YOUR_API_KEY_HERE")
client = Groq(api_key=api_key)

LLM_MODEL = "llama-3.1-8b-instant"
```

```python
CATEGORY_DEFINITIONS = """
You must classify NON-TOXIC comments into exactly ONE of these three categories:

1. non_sexual:
   Neutral comments mentioning gender/sex related words, but not sexual content.

2. motivational_non_hate:
   Comments expressing gratitude, support, encouragement, optimism, praise.

3. non_abusive_violence:
   Neutral descriptions of violence, war, crime, fighting, or conflict without insults or threats.

Output must be ONLY a JSON list of objects:
[
  {"subcategory": "..."},
  {"subcategory": "..."},
  ...
]
"""

def build_batch_prompt(text_list):
    items = ""
    for i, t in enumerate(text_list):
        items += f"TEXT_{i}: \"\"\"{t}\"\"\"\n\n"

    return f"""
{CATEGORY_DEFINITIONS}

Classify the following texts:

{items}

Return ONLY JSON list with exactly one subcategory for each TEXT_i.
"""
```

```python
import time

MAX_REQUESTS_PER_MIN = 5  # stay well below 30 RPM
SECONDS_PER_REQUEST = 60.0 / MAX_REQUESTS_PER_MIN  # 12 seconds per request
```

```python
VALID_SUBCATS = ["non_sexual", "motivational_non_hate", "non_abusive_violence"]

def annotate_batch(text_batch, max_retries=3):
    """
    text_batch: list of strings
    returns: (labels_list, raw_response_string)
    labels_list is ALWAYS the same length as text_batch.
    """
    prompt = build_batch_prompt(text_batch)

    for attempt in range(max_retries):
        try:
            response = client.chat.completions.create(
                model=LLM_MODEL,
                messages=[
                    {"role": "system", "content": "You are an annotation assistant."},
                    {"role": "user", "content": prompt}
                ],
                temperature=0.0,
                max_tokens=128,  # keep output short to respect TPM
            )

            raw = response.choices[0].message.content

            # Initialize labels with None for each input text
            labels = [None] * len(text_batch)

            # Try strict JSON parsing
            try:
                data = json.loads(raw)
                if isinstance(data, list):
                    for i, item in enumerate(data):
                        if i >= len(labels):
                            break
                        lab = item.get("subcategory")
                        if lab in VALID_SUBCATS:
                            labels[i] = lab
                # any remaining labels stay None
            except Exception:
                # JSON parse failed; leave labels as None for now
                pass

            # If all None, apply a very crude heuristic
            if all(l is None for l in labels):
                for i in range(len(labels)):
                    for cand in VALID_SUBCATS:
                        if cand in raw:
                            labels[i] = cand
                            break

            # Ensure labels length matches text_batch length
            if len(labels) != len(text_batch):
                if len(labels) < len(text_batch):
                    labels += [None] * (len(text_batch) - len(labels))
                else:
                    labels = labels[:len(text_batch)]

            return labels, raw

        except Exception as e:
            wait = 5 * (attempt + 1)
            print(f"Error in annotate_batch (attempt {attempt+1}/{max_retries}): {e}")
            print(f"Sleeping {wait} seconds before retrying...")
            time.sleep(wait)

    # All retries failed
    return [None] * len(text_batch), None
```

```python
from tqdm import tqdm
import pandas as pd

BATCH_SIZE = 5
progress_path = "data/llm_annotation_progress.csv"

# Always load the latest progress from disk before continuing
progress_df = pd.read_csv(progress_path)

for start_idx in tqdm(range(0, len(progress_df), BATCH_SIZE)):

    batch = progress_df.iloc[start_idx:start_idx + BATCH_SIZE]

    # Skip batch if all already annotated
    if batch["llm_pred_category"].notna().all():
        continue

    texts = batch["text"].tolist()
    labels, raw = annotate_batch(texts)

    # Safety: ensure we have a label per text
    if len(labels) != len(texts):
        print(f"Warning: labels length {len(labels)} != texts length {len(texts)} at start_idx {start_idx}")
        if len(labels) < len(texts):
            labels += [None] * (len(texts) - len(labels))
        else:
            labels = labels[:len(texts)]

    for i, idx in enumerate(batch.index):
        progress_df.at[idx, "llm_pred_category"] = labels[i]
        progress_df.at[idx, "llm_raw_response"] = raw

    # Save checkpoint every 100 rows (20 batches)
    if start_idx % (BATCH_SIZE * 20) == 0:
        progress_df.to_csv(progress_path, index=False)
        print(f"Checkpoint saved at index {start_idx}")

    # Throttle to respect RPM and TPM: at most 5 requests per minute
    time.sleep(SECONDS_PER_REQUEST)

# Final save
progress_df.to_csv(progress_path, index=False)
print("Final progress saved to", progress_path)
```

```python
def normalize_text(text: str) -> str:
    if pd.isna(text):
        return ""

    # Safe cast
    text = str(text)

    # Fix mojibake and weird encodings
    text = fix_text(text)

    # Replace explicit tokens
    text = text.replace("NEWLINE_TOKEN", " ")

    # Normalize newlines to spaces
    text = text.replace("\r", " ").replace("\n", " ")

    # Collapse repeated quotes
    text = re.sub(r'"+', '"', text)
    text = re.sub(r"'+", "'", text)

    # Remove repeated spaces
    text = re.sub(r"\s+", " ", text)

    # Strip leading/trailing quotes and spaces
    text = text.strip(" '\"\t")

    return text.strip()
```

```python
import string

def is_useful_text(text: str) -> bool:
    if not text:
        return False

    t = text.strip()
    if not t:
        return False

    # Only punctuation or quotes
    if all(ch in string.punctuation for ch in t):
        return False

    # Very short overall (e.g. 1–4 chars) and no spaces
    if len(t) <= 3 and " " not in t:
        # Allow some short strongly meaningful tokens if you want
        keep_tokens = {"wtf", "lol", "ffs"}  # can tweak
        if t.lower() not in keep_tokens:
            return False

    # Require at least a few alphabetic characters
    alpha_count = sum(ch.isalpha() for ch in t)
    if alpha_count < 3:
        return False

    return True
```

```python
df_clean = df.copy()

# Normalize text
df_clean["text"] = df_clean["text"].astype(str).apply(normalize_text)

# Compute mask of useful rows
useful_mask = df_clean["text"].apply(is_useful_text)

dropped = (~useful_mask).sum()
kept = useful_mask.sum()

print("Total rows:", len(df_clean))
print("Kept rows:", kept)
print("Dropped rows as useless:", dropped)

df_clean = df_clean[useful_mask].reset_index(drop=True)

print("\nAfter cleaning shape:", df_clean.shape)
print("Final category distribution after cleaning:")
print(df_clean["final_category"].value_counts())

# Save cleaned single-file dataset
output_path = "data/final_10k_dataset_cleaned.csv"
df_clean.to_csv(output_path, index=False)
print("\nCleaned dataset saved to:", output_path)
```
